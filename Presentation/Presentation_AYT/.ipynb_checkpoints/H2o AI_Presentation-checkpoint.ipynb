# H2O AI for Model Selection and Prediction

```python
import os
import h2o
import numpy as np
import pandas as pd

from h2o.automl import H2OAutoML
from sklearn.model_selection import train_test_split

%matplotlib inline
```

### We first need to initialize the H2o cluster to begin operations

```python
h2o.init()
```

# Regression : House Prices Prediction

Link to Kaggle Competition : https://www.kaggle.com/c/house-prices-advanced-regression-techniques

Importing the dataset into H2O data frames

```python
house_train_df = h2o.import_file('House_train.csv')
```

```python
house_train_df.head()
```

```python
house_train_df.describe()
```

```python
house_test_df = h2o.import_file('House_test.csv')
```

Here we set the **max_run_time**, which basically sets an upper limit on the number of seconds that the H2O AutoML should run before returning its leaderboard

```python
run_time_secs = 240

aml_reg = H2OAutoML(max_runtime_secs = run_time_secs, project_name = 'House_Comparison')
```

```python
X_reg = house_train_df.columns
y_reg = 'SalePrice'
X_reg.remove(y_reg)
#Since ID is redundant
X_reg.remove('Id')
```

```python
aml_reg.train(x = X_reg, y = y_reg, training_frame = house_train_df)
```

```python
aml_reg.leaderboard.as_data_frame()
```

```python
aml_reg.leader.varimp_plot()
```

```python
aml_reg.leader.partial_plot(data=house_train_df,cols=['OverallQual','GrLivArea','1stFlrSF','GarageCars'],nbins=20)
```

```python
h2o_preds = aml_reg.predict(test_data = house_test_df)
```

```python
sub = house_test_df.cbind(h2o_preds)
sub = sub[['Id','predict']]
sub.set_names(['Id','SalePrice'])
```

```python
h2o.export_file(sub[['Id','SalePrice']],os.getcwd()+'\\'+'House_Submission1.csv')
```

# Classification : Titanic Survived or Not Classification

Link for Kaggle Competition : https://www.kaggle.com/c/titanic

```python
titanic_train_df = h2o.import_file('titanic_train.csv')
```

```python
titanic_test_df = h2o.import_file('titanic_test.csv')
```

```python
run_time_secs = 240

aml_cl = H2OAutoML(max_runtime_secs = run_time_secs, project_name = 'Titanic_Comparison')
```

```python
titanic_train_df.describe()
```

Since the target variable **Survived** is a categorical variable that has been encoded as an Integer, we need to inform H2O that it is a category by calling the asfactor() method on that column.

```python
X_class = titanic_train_df.columns
y_class = "Survived"
X_class.remove(y_class)

#Removing all redundant variables
X_class.remove('PassengerId')
X_class.remove('Name')
X_class.remove('Ticket')
```

```python
titanic_train_df[y_class] = titanic_train_df[y_class].asfactor()
```

```python
aml_cl.train(x = X_class, y = y_class, training_frame = titanic_train_df)
```

```python
aml_cl.leaderboard.as_data_frame()
```

```python
pred_cl = aml_cl.predict(titanic_test_df)
```

```python
sub_cl = titanic_test_df.cbind(pred_cl)
```

```python
sub_cl = sub_cl[['PassengerId','predict']]
sub_cl.set_names(['PassengerId','Survived'])
```

```python
h2o.export_file(sub_cl,os.getcwd()+'\\'+'Titanic_Submission.csv')
```
